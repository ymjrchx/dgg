```python
# 对product_amount字段进行0,1转换,缺失0，未成单，负样本，有值为1，成单，正样本
def nan_encoder(data,variances):
    df = data.loc[:,variances]
    df = df.notnull()*1
    columns = [i+"is_nan" for i in variances]
    df.columns = columns
    return df
```

```python
# 对orderAmount,mon,times字段进行0,1转换,
# orderAmount：订单金额--是否有贷款，0表示没有订单金额，1表示有
# mon:最近一次贷款时长--是否有贷款，0表示没有贷款，1表示有
# times：最近90天备注次数--是否有备注，0表示没有，1表示有
data_2 = nan_encoder(data, ['product_amount','refundOrderSign','signOderNum'])
data = pd.concat([data, data_2],axis = 1)
```

# 样本均衡处理

## 负样本下采样

```python
def lower_sample_data(df, variance, percent = 3):   
    '''    percent:多数类别下采样的数量相对于少数类别样本数量的比例    '''   
    data1 = df[df[variance] == 0]  # 将多数类别的样本放在data1   
    data0 = df[df[variance] == 1]  # 将少数类别的样本放在data0  
    index = np.random.randint(        
        len(data1), size = percent * (len(df) - len(data1)))  # 随机给定下采样取出样本的序号 
    lower_data1 = data1.iloc[list(index)]  # 下采样  
    return(pd.concat([lower_data1, data0]))
```
