# 1.降维
1. 当特征选择完成后，可以直接训练模型了，但是可能由于特征矩阵过大，导致计算量大，训练时间长的问题，因此降低特征矩阵维度也是必不可少的
2. 降维方法:
> 1. 基于L1惩罚项的模型
> 2. 主成分分析法（PCA）
> 3. 线性判别分析（LDA），线性判别分析本身也是一个分类模型
3. PCA和LDA
> 1. 相似点:本质是要将原始的样本映射到维度更低的样本空间中
> 2. PCA和LDA的映射目标不一样：PCA是为了让映射后的样本具有最大的发散性；而LDA是为了让映射后的样本有最好的分类性能。
> 3. 所以说PCA是一种无监督的降维方法，而LDA是一种有监督的降维方法。

# 2.PCA

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
iris = load_iris()
#主成分分析法，返回降维后的数据
#参数n_components为主成分数目

# 1.简单使用
PCA(n_components=2).fit_transform(iris.data)
```

```python
# 2.使用案例
from sklearn.decomposition import PCA,KernelPCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris_data = load_iris()
# 1.将标签进行量化，就是说本来都是字符串啊，但是最后计算的时候都需要量化成1，2，3类等
category = pd.Categorical(iris_data.target) 
# 2.使用PCA降到2维度
pca_2c = PCA(n_components=2)

x_pca_2c = pca_2c.fit_transform(iris.data)
print(x_pca_2c.shape)
plt.scatter(x_pca_2c[:,0],x_pca_2c[:,1],c=category.codes)
plt.show()
```

# 3.LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
#线性判别分析法，返回降维后的数据
#参数n_components为降维后的维数

# 1.简单使用
LDA(n_components=2).fit_transform(iris.data, iris.target)
```

```python
# 2.使用案例
#LDA相较于pca是有监督的，但不能用于回归分析
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris_data = load_iris()
category = pd.Categorical(iris.target)

lda_2c = LDA(n_components=2)
x_lad_2c = lda_2c.fit_transform(iris_data.data,iris_data.target)
print(x_lad_2c .shape)
plt.scatter(x_lad_2c[:,0],x_lad_2c[:,1],c=category.codes)
plt.show()
```
