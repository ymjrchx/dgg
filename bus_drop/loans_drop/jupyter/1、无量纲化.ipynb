```python
from sklearn.datasets import load_iris
import sklearn
```

# 1. 导入数据集

```python
# 实例化
iris = load_iris()
# 特征矩阵
data = iris.data
data
```

```python
# 目标变量
target = iris.target
target
```

# 2. 标准化

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard = StandardScaler().fit_transform(data)
standard
```

# 3、特征缩放

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler().fit_transform(data)
minmax
```

# 4、定量特征二值化

```python
from sklearn.preprocessing import Binarizer
```

```python
binarizer = Binarizer(threshold=2).fit_transform(data)
binarizer
```

# 5、定性特征哑变量编码--onehot

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder().fit_transform(target.reshape(-1,1))
onehot
```
