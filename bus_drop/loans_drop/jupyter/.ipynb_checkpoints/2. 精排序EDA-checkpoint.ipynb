## 2.3 缺失值占比分析

```python
def null_rate(data):
    """得到每列缺失值比例的数据框"""
    null_rate = pd.DataFrame(1-data.count()/data.shape[0]).reset_index()
    null_rate.columns = ["feature_name","null_rate"]
    null_rate = null_rate.sort_values(by="null_rate",ascending=False)
    return null_rate
```

```python
null_rate1 = pd.DataFrame(1-data.count()/data.shape[0]).reset_index() #缺失的比例
# null_rate = data.isnull().mean()
null_rate1.columns = ["feature_name","null_rate"]
null_rate = null_rate1.sort_values(by="null_rate",ascending=False)
```

```python
null_rate[null_rate["feature_name"]=="originCode"]
```

```python
threshold = 0.8
print("data数据中缺失率低于%s的特征个数为%s"%(threshold,(null_rate["null_rate"]>threshold).sum()))
```

```python
def filter_null_rate(data,threshold):
    """过滤缺失率小于阈值的特征"""
    df = data.loc[data["null_rate"] <= threshold,:]
    print("缺失率小于%s的特征有%s个"%(threshold,df.shape[0]))
    return df
```

```python
def drop_mul_col(df,drop_list,inplace=True):
    """根据要删除的列名，一次性删除多列"""
    columns = list(df.columns)
    print("原数据框一共有%s列"%len(columns))
    col_index = [columns.index(i) for i in drop_list]
    data = df.drop(df.columns[col_index],axis = 1,inplace=True)
    print("数据框字段删除完毕,剩余%s列"%len(data.columns))
    return data
```
