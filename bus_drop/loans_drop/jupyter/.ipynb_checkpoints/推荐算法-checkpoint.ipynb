# 标准化 归一化

```python
from sklearn import  preprocessing

std_scale = preprocessing.StandardScaler().fit(business_df[['芝麻信用分']])
print(std_scale)
business_std = std_scale.transform(business_df[['芝麻信用分']])
print(business_std)
print('Mean after standardization:\n客户={:.2f}'
      .format(business_std[:,0].mean()))
print('\nStandard deviation after standardization:\n客户={:.2f}'
      .format(business_std[:,0].std()))

business_df.fillna(business_df.mean(),inplace=True)
std_scale = preprocessing.StandardScaler().fit(business_df[['房产市价']])
print(std_scale)
house_std = std_scale.transform(business_df[['房产市价']])
print(house_std)
print('Mean after standardization:\n客户={:.2f}'
      .format(house_std[:,0].mean()))
print('\nStandard deviation after standardization:\n客户={:.2f}'
      .format(house_std[:,0].std()))
```

```python
minmax_scale = preprocessing.MinMaxScaler().fit(salesMan_df[['芝麻信用分']])
print(minmax_scale)
salesMan_scale = minmax_scale.transform(salesMan_df[['芝麻信用分']])
print(salesMan_scale)
print('Min-value after min-max scaling:\nAlcohol={:.2f}'
      .format(salesMan_scale[:,0].min()))
print('\nMax-value after min-max scaling:\nAlcohol={:.2f}'
      .format(salesMan_scale[:,0].max()))

salesMan_df.fillna(business_df.mean(),inplace=True)
minmax_scale = preprocessing.MinMaxScaler().fit(salesMan_df[['操行分']])
print(minmax_scale)
score_scale = minmax_scale.transform(salesMan_df[['操行分']])
print(salesMan_scale)
print('Min-value after min-max scaling:\nAlcohol={:.2f}'
      .format(score_scale[:,0].min()))
print('\nMax-value after min-max scaling:\nAlcohol={:.2f}'
      .format(score_scale[:,0].max()))
```

```python
record = np.array(record,dtype=int)
record
```

```python
def normalizeRatings(rating, record):
    m,n = rating.shape
    rating_mean = np.zeros((m,1))
    rating_norm = np.zeros((m,n))
    for i in range(m):
        idx = record[i,:] != 0
        print(idx, rating[i,idx])
        rating_mean[i] = np.mean(rating[i,idx])
        rating_norm[i, idx] -= rating_mean[i]
    return rating_norm, rating_mean

rating_norm, rating_mean = normalizeRatings(rating, record)
print(rating_norm)
print(rating_mean)
```
