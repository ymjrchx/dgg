## 处理label值

```python
# 删除'call_duration', 'follow_days', 'product_amount'都是缺失的值
total_data.dropna(subset=['call_duration', 'follow_days', 'product_amount'], axis=0, how='all', inplace=True)
```

```python
# 对product_amount字段进行0,1转换,缺失0，未成单，负样本，有值为1，成单，正样本
def nan_encoder(data,variances):
    df = data.loc[:,variances]
    df = df.notnull()*1
    columns = [i+"is_nan" for i in variances]
    df.columns = columns
    return df
```

```python
data_1 = nan_encoder(total_data, ['product_amount'])
total_data = pd.concat([total_data, data_1],axis = 1)
total_data.rename(columns={"product_amountis_nan": "label"},inplace = True)
```

```python
#字段iscost：yes,jscb是‘1’，其余为0
def tran(row):
    if row['isCost'] == 'yes':
        return 1
    elif row['isCost'] == 'JSCB':
        return 1
    elif row['isCost'] == 1:
        return 1 
    else:
        return 0
data['isCost'] = data.apply(tran,axis=1)
```

```python
#字段rank：缺失替换为0，P替换为1，P0替换为2，L替换为3
def tran(row):
    if row['rank'] == 'P':
        return 1
    elif row['rank'] == 'P0':
        return 2
    elif row['rank'] == 'L':
        return 3
    else:
        return 0
data['rank'] = data.apply(tran,axis=1)
```

```python
#字段post：缺失替换为0，P替换为1，P0替换为2，L替换为3
def tran(row):
    if row['post'] == '商务顾问':
        return 1
    elif row['post'] == '电销专员':
        return 0
    else:
        return 2
data['post'] = data.apply(tran,axis=1)
```

```python
# 离散型变量,如果某一取值的数量超过了95%，则舍弃该变量
remove = set()
print('处理类别型变量')
for k in category_columns+bool_columns:
    counts = data[k].value_counts()
    num = sum(counts)
    for v in counts:
        if v/num>=0.95:
#             print(k)
#             print(counts)
            remove.add(k)
print('处理连续型变量')
for k in continues_columns:
    describe = data[k].describe()
    if describe[4]==describe[6]:
        remove.add(k)
#         print(k,describe[4],describe[6])
remove = list(remove)
remove
```

```python
#1. 进行分箱得到切分点的列表，并持久化
cutoffs_dict = dict()
for i in continues_columns:
    bin_chi = BinChi()
    x = '%s'%i
    y = 'label'
    cutoffs = bin_chi.chi_merge(df=data, x=x, y=y, max_groups=None)
    cutoffs_dict.setdefault(i,0)
    cutoffs_dict[i] = cutoffs
```

```python
def nan_encoder(data,variances):
    df = data.loc[:,variances]
    df = df.notnull()*1
    columns = [i+"is_nan" for i in variances]
    df.columns = columns
    return df
data_3 = nan_encoder(x_test, ['orderAmount','mon','times'])
x_test = pd.concat([x_test, data_3],axis = 1)
x_test.drop(columns = ['orderAmount','meetingCount', 'orderCount',
                     'tccs','mon','remarkType','times', 'businessStage',
                     'extensionCode'], axis = 1, inplace=True)
x_test.fillna(value = {'call_duration':0,'follow_days':0,'salary_score':0,
                     'followTimes':0,'allotNumber':0},inplace = True)
# coverCharge历史总服务费用,缺失为0；signOderNum，历史签单数量，缺失为0；
x_test.fillna(value = {'coverCharge':0, 'signOderNum':0},inplace = True) 
# 'referral','vip','isMind',变量数据是0,1 ，缺失值用2填充，代表未知；
# refundOrderSign是否退单  0为未退单，1为退单，新增类型2，代表未知
x_test.fillna(value = {'referral':2,'vip':2, 'refundOrderSign':2,'isMind':2},inplace = True)
def tran(row):
    if row['isCost'] == 'yes':
        return 1
    elif row['isCost'] == 'JSCB':
        return 1
    else:
        return 0
x_test['isCost'] = x_test.apply(tran,axis=1)
def tran(row):
    if row['rank'] == 'P':
        return 1
    elif row['rank'] == 'P0':
        return 2
    elif row['rank'] == 'L':
        return 3
    else:
        return 0
x_test['rank'] = x_test.apply(tran,axis=1)
def tran(row):
    if row['post'] == '商务顾问':
        return 1
    elif row['post'] == '电销专员':
        return 0
    else:
        return 2
x_test['post'] = x_test.apply(tran,axis=1)
# 非数值型变量用众数填充
x_test.fillna(value = mode_miss,inplace = True)  
# 偏态分布，用中位数填充
x_test.fillna(value = median_miss,inplace = True)
# 正态分布，均值填充
x_test.fillna(value = mean_miss,inplace = True)
```
