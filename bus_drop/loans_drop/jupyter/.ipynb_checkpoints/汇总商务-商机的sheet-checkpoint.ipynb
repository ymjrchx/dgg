```python
import pandas as pd
import numpy as np
import tensorflow as tf
import random
from tqdm import tqdm_notebook as tqdm
```

```python
follow_df = pd.read_excel('跟进日期3.1-3.10.xls',encoding = 'utf-8')
print(follow_df.shape)
print(follow_df.info())
follow_df.head()
```

```python
sheet_names = ['Sheet1(2)','Sheet1(3)','Sheet1(4)','Sheet1(5)','Sheet1(6)','Sheet1(7)']
column_names = follow_df.columns.values.tolist()
dfs = [follow_df]
for i in tqdm(range(len(sheet_names))):
    df2 = pd.read_excel('跟进日期3.1-3.10.xls',encoding = 'utf-8',sheet_name=sheet_names[i],header=None)
    df2.columns = column_names
    print(df2.shape)
    dfs.append(df2)

df = pd.concat(dfs)
print(df.shape)
print(df.info())
df.head()
```

```python
writer = pd.ExcelWriter('跟单记录3.1-3.10.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
