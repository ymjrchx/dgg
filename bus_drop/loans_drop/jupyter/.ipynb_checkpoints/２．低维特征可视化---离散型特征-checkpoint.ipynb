```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
import seaborn as sns
sns.set(style="whitegrid",color_codes=True)
%matplotlib inline
```

```python
np.random.seed(sum(map(ord,"categorical")))
```

```python
#　导入数据
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
tips.head()
```

# 1. 类别型变量和其他变量大致关系
1. 只要其中一个变量是类别型变量就行，不是离散型变量都可以的
2. 工具：①sns.stripplot函数 ②sns.swarmplot函数
3. 缺点：无法某一个类别中对应的某个变量的准确分布　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　

## １.1．sns.stripplot函数,条带图
> 1. 作用:按照ｘ属性对应的类别，分别展示ｙ的值．
> 2. 用于离散型和其他任意类型的变量,但是ｘ，ｙ必须有一个是数值型
> 3. 参数：jitter=True,当数据点重合过多时，一个点附近有大量的样本，这时覆盖太严重了．可用该参数调整
> 4. 参数：dodge = True,控制组内分类是否彻底分拆

```python
# 1. 类别型和连续型变量关系
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.stripplot(x="day",y="size",data=tips)
```

```python
# 2. 拆分覆盖严重的点
sns.stripplot(x="day",y="total_bill",data=tips,jitter = True)
```

```python
# 3.控制组内分类彻底分拆
sns.stripplot(x="time",y="total_bill",data=tips,hue = "day",jitter =True,dodge=True)
```

```python
# 4. 参数order,对ｘ参数所选字段内的类别进行排序和筛选
sns.stripplot(x="time",y="total_bill",data=tips,jitter =True,hue="day",
             dodge = True,order =["Dinner","Lunch"])
# sns.stripplot(x="time",y="total_bill",data=tips,jitter =True,hue="day",
#              dodge = True,order =["Dinner"])
```

## 1.2. sns.swarmplot,蜂群图
> 1. 作用，类似于加jitter=True参数的sns.stripplot函数，拆分重复过多的点
> 2. 同时试用于离散型与离散型／连续型变量
> 3. 缺点：计算耗时

```python
sns.swarmplot(x="day",y="total_bill",data = tips)
```

```python
# sns.swarmplot对于类别型和类别型之间效果更明显
ok,[ax0,ax1] = plt.subplots(nrows=2,ncols=1,figsize = [12,8])
sns.stripplot(x="size", y="size", data=tips,jitter= True, ax = ax0);
sns.swarmplot(x="size", y="size", data=tips,ax = ax1)
```

#  2.离散型变量对应的特征分布
1. 作用：查看某个类别对应的某个特征的分布,上面的类别型数据的可视化可能会无法反应某类中信息的一个分布情况，尤其是两个类别中总数相近的时候
2. 工具：sns.boxplot＼sns.violinplot函数

## 2.1. sns.boxplot－－－箱线图
1. hue添加第三维度，并默认dodge躲避＝Ｔｒｕｅ，达到第三维度分割的目的

```python
sns.boxplot(x="day",y="total_bill",data = tips)
```

```python
# hue---第三维度，在箱线图中会被分割出来，形成多个小的箱线图,可以用dodge　＝　False控制
sns.boxplot(x="day",y="total_bill",hue="size",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",hue="size",data=tips,dodge=False)
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.boxplot(x="day", y="total_bill", hue="weekend", data=tips, dodge=False)
```

## 2.2. sns.violinplot小提琴图

```python
sns.violinplot(x="day", y="total_bill", hue="time", split=True, data=tips)
```

Violinplot 与 Swarmplot
其实我们发现这个函数和我们的swarmplot很类似,但是violinplot能让我们一眼就识别出数据的分布情况.非常清晰直观

```python
sns.swarmplot(x = 'day', y ='total_bill',hue="time",data = tips)
```

sns.violinplot与sns.swarmplot合并

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="w", alpha=.5)
```

# 3. 类别特征的统计信息
1. 作用：不仅仅查看数据的外在表现如分布，还要查看数据的内在信息，如统计信息
2. 工具：sns.barplot

## 3.1. sns.barplot直方图
> 1. 默认计算为１的比例
> 2. hue添加第三维度

```python
sns.barplot(x="sex",y = "survived",data=titanic)
```

```python
sns.barplot(x="sex",y="survived",hue="class",data=titanic)
```

## 3.2. sns.countplot---查看个数
> 1. countplot函数不能同时使用x,y所以如果想要统计某个cate变量对应的变量的个数最好用hue进行分开
> 2. countplot函数ｙ为count,计数

```python
sns.countplot(hue="sex",x="survived",data=titanic)
```

## 3.3. sns.pointplot变形的折线图,连接第三维度
1. 类似于直方图，ｙ为计算对应的概率，
2. 但是会对相同的hue特征进行连接，得到特征值变化的曲线

```python
sns.pointplot(x="sex",y="survived",hue="class",data=titanic)
```

```python
sns.pointplot(x="sex", y="survived", hue="class", data=titanic,
              palette={"First": "g", "Second": "m", "Third":'b'},
              markers=["^", "o","+"], linestyles=["-", "--",""])
```

```python
# orient="h"表示按横轴绘制
sns.boxplot(data=iris, orient="h")
```

```python
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c")
```

## 3.4.factorplot因子图,默认折线图
> 1. 可以有三个维度，两个类别，一个连续的关系图，默认折线图
> 2. 可以通过kind＝＂ｂａｒ＂,将折线图修改为直方图
> 3. 通过ｃｏｌ参数添加到４个维度，三个离散，一个连续

```python
sns.factorplot(x="day",y="total_bill",hue="smoker",data=tips)
```

```python
sns.factorplot(x="day",y="total_bill",hue="smoker",kind="bar",data=tips)
```

```python
sns.barplot(x="day",y="total_bill",hue="smoker",data=tips)
```

```python
sns.factorplot(x="day",y="total_bill",hue="smoker",col="time",kind="bar",data=tips)
```

```python
sns.factorplot(x="day", y="total_bill", hue="smoker", col="time", 
               data=tips, kind="box", size=4, aspect=.5)
```
