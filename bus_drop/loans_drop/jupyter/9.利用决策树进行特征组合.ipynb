# 1.labelencoder将字符串\类别变量变成数值变量

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
print(data.shape)
data.head()
```
