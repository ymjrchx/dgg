# 2. 牛顿插值法

```python
import matplotlib.pyplot as plt
from pylab import mpl
import math
"""
牛顿插值法
插值的函数表为
xi      0.4，       0.55，     0.65，      0.80，       0.90，   1.05
f(xi)   0.41075,    0.57815,   0.69675,    0.88811,    1.02652,  1.25382
"""
 
x = [0.4, 0.55, 0.65, 0.80, 0.90, 1.05]
y = [0.41075, 0.57815, 0.69675, 0.88811, 1.02652, 1.25382]
 
"""计算五次差商的值"""
def five_order_difference_quotient(x, y):
    # i记录计算差商的次数，这里循环5次，计算5次差商。
    i = 0
    quotient = [0, 0, 0, 0, 0, 0]
    while i < 5:
        j = 5
        while j > i:
            if i == 0:
                quotient[j]=((y[j]-y[j-1])/(x[j]-x[j-1]))
            else:
                quotient[j] = (quotient[j]-quotient[j-1])/(x[j]-x[j-1-i])
            j -= 1
        i += 1
    return quotient;
 
def function(data):
    return x[0]+parameters[1]*(data-0.4)+parameters[2]*(data-0.4)*(data-0.55)+\
           parameters[3]*(data-0.4)*(data-0.55)*(data-0.65)\
           +parameters[4]*(data-0.4)*(data-0.55)*(data-0.80)
 
"""计算插值多项式的值和相应的误差"""
def calculate_data(x,parameters):
    returnData=[];
    for data in x:
        returnData.append(function(data))
    return returnData
 
"""画函数的图像
newData为曲线拟合后的曲线
"""
 
def draw(newData):
    plt.scatter(x,y,label="离散数据",color="red")
    plt.plot(x,newData,label="牛顿插值拟合曲线",color="black")
    plt.scatter(0.596,function(0.596),label="预测函数点",color="blue")
    plt.title("牛顿插值法")
    mpl.rcParams['font.sans-serif'] = ['SimHei']
    mpl.rcParams['axes.unicode_minus'] = False
    plt.legend(loc="upper left")
    plt.show()
 
 
parameters=five_order_difference_quotient(x,y)
yuanzu=calculate_data(x,parameters)
draw(yuanzu)
```

```python
def tree2bin(df,Kvar,Yvar,max_depth,p):
    """决策树分箱,dataframe为数据框,Kvar为待分箱变量,Yvar目标变量,p决策树叶子节点数据占比"""
    df_bin = df[[Kvar,Yvar]]
    df_Xvar = df.drop([Kvar,Yvar],axis = 1)
    dict_bin ={}
    from sklearn import tree
    for col in df_Xvar.columns: #每个想分箱的自变量
        cur_mob_dur = tree.DecisionTreeClassifier(criterion="entropy",
                                                  max_depth = max_depth,min_samples_leaf = int(p*len(df)))
        cur_mob_dur = cur_mob_dur.fit(df[col].values.reshape(-1,1),df[Yvar].values.reshape(-1,1)) #单变量决策树
        cut_point = cur_mob_dur.tree_.threshold[cur_mob_dur.tree_.threshold!=-2] #得到连续变量分割点
        
        Nvar = np.zeros(df[col].shape) #用于存放
        inter_list =[]
        if len(cut_point) ==1:
            Nvar[np.array(df[col] < cut_point[0])] =1
            Nvar[np.array(df[col] >= cut_point[0])] =2
            inter_list = [[1,-100000000,cut_point[0]],[2,cut_point[0],100000000]]
        elif len(cut_point) > 1:
            cut_point.sort()
            Nvar[np.array(df[col]<cut_point[0])] =1
            inter_list = [[1,-100000000,cut_point[0]]]
            for i in range(len(cut_point)-1):
                Nvar[np.array((df[col] >=cut_point[i]) & (df[col] < cut_point[i+1]))] = i+2
                inter_list = inter_list + [[i+2,cut_point[i]]]  
            Nvar[np.array(df[col] >= cut_point[len(cut_point)-1])] = len(cut_point)+1
            inter_list = inter_list+[[len(cut_point)+1,cut_point[len(cut_point)-1],100000000]]
        else :
            Nvar = 1
            inter_list = np.array([1,-100000000,100000000]).reshape(1,-1)
            
        # 分箱结果添加到原数据框
        df_bin[col] = Nvar
        # 
        inter_df = pd.DataFrame(inter_df)
        inter_df.columns =['bin',"lower","upper"]
        # bad rate
        cross = pd.crosstab(df_bin[col],df_bin[Yvar])
        cross.columns = ["Good","Bad"]
        cross["All"] = cross["Good"] + cross["Bad"] 
        cross["Percent"] = cross.All/sum(cross.All) 
        cross["Bad_Rate"] = cross.Bad/cross.All
        inter_df =  pd.merge(inter_df,cross,left_on ="bin",right_index=True)
        dict_bin[col] = inter_df
    return df_bin,dict_bin
```
