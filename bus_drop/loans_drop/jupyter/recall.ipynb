```python
train = dict()  # 用户-物品的评分矩阵
for line in open('train_data.csv',encoding='utf-8'):
    user, item, score  = line.strip().split(",")
    train.setdefault(user, {})
    train[user][item] = score
```

```python
# 建立物品-物品的共现矩阵
C = dict()  # 物品-物品的共现矩阵
N = dict()  # 物品被多少个不同用户购买
for user, items in train.items():
    for i in items.keys():
        N.setdefault(i, 0)
        N[i] += 1
        C.setdefault(i, {})
        for j in items.keys():
            if i == j: continue
            C[i].setdefault(j, 0)
            C[i][j] += 1
            # 计算相似度矩阵
```

```python
import math
W = dict()
for i, related_items in C.items():
    W.setdefault(i, {})
    for j, cij in related_items.items():
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
```

```python
def recommend(user, k = 10, nitem = 5):
    rank = dict()
    ru = train.get(user,{})
    for item,score in ru.items():
        try:
            for j,wj in sorted(W[item].items(), key = lambda x:x[1], reverse = True)[0:k]:
                if j in ru:
                    continue
                rank.setdefault(j, 0)
                rank[j] += float(score) * wj
        except KeyError:
            continue  
    return dict(sorted(rank.items(), key = lambda x:x[1], reverse = True)[0:nitem])
```

```python
def predict(user, k = 10, nitem = 5):
    pre = dict()
    similar = dict()
    
    ru = train.get(user,{})
    for item, score in ru.items():
        topk = sorted(W[item].items(),key = lambda x : x[1], reverse = True)[0:k]
        for j, wj in topk:
            if j in ru:
                continue
            pre.setdefault(j, 0)
            similar.setdefault(j, 0)
            pre[j] += float(score) * wj
            similar[j] += wj
    rank = dict()
    for a, b in pre.items():
        if similar.get(a):
            rank.setdefault(a, 0)
            rank[a] = float(pre[a]/similar[a])
    final = dict(sorted(rank.items(), key = lambda x:x[1], reverse = True)[0:nitem])
    return pre,similar,rank    #final
```

```python
# -*- coding: utf-8 -*-  

 
import math
import random
import os 
from itertools import islice #islice 返回迭代器
 
class ItemCF:
    def __init__(self, datafile = None):
        self.datafile = datafile
        self.readData()
        self.splitData()
      
    def readData(self,datafile = None):
        self.datafile = datafile or self.datafile
        self.data = []
        file = open(self.datafile,'r')
        for line in islice(file, 1, None): #file.readlines(): 读取记录
            userid, itemid, record = line.strip().split(',') #使用逗号作为分隔符
            self.data.append((userid,itemid,float(record)))
                              
    def splitData(self,data=None,prob=0.2,seed=10):
        self.traindata = {}
        self.testdata = {}
        self.alldata = {}
        data = data or self.data
        random.seed(seed)#生成随机数
        for user,item,record in self.data:
            self.alldata.setdefault(user,{})
            self.alldata[user][item] = record
            
            if random.random() <= prob:
                self.traindata.setdefault(user,{})
                self.traindata[user][item] = record  
                
        for user,item,record in self.data:
            if user in self.traindata.keys():
                if item not in self.traindata[user].keys():
                    self.testdata.setdefault(user,{})
                    self.testdata[user][item]= record
                else:
                    continue
            else:
                continue
                                  
    def ItemSimilarity(self, train = None):
        train = train or self.traindata
        self.itemSim = dict()
        item_user_count = dict() #item_user_count{item: likeCount} the number of users who like the item
        count = dict() #count{i:{j:value}} the number of users who both like item i and j
        for user,item in train.items(): #initialize the user_items{user: items}
            for i in item.keys():
                item_user_count.setdefault(i,0)
                item_user_count[i] += 1
                for j in item.keys():
                    if i == j:
                        continue
                        
                    count.setdefault(i,{})
                    count[i].setdefault(j,0)
                    count[i][j] += 1
        for i, related_items in count.items():
            self.itemSim.setdefault(i,dict())
            for j, cuv in related_items.items():
                self.itemSim[i].setdefault(j,0)
                self.itemSim[i][j] = cuv / math.sqrt(item_user_count[i] * item_user_count[j] * 1.0)
                              
    def recommend(self,user,train = None, k = 10,nitem = 5):
        train = train or self.traindata
        rank = dict()
        ru = train.get(user,{})
        for i,pi in ru.items():
            try:
                for j,wj in sorted(self.itemSim[i].items(), key = lambda x:x[1], reverse = True)[0:k]:
                    if j in ru:
                        continue
                    rank.setdefault(j,0)
                    rank[j] += pi*wj
            except KeyError:
                continue  
        return dict(sorted(rank.items(), key = lambda x:x[1], reverse = True)[0:nitem])
              
    def recallAndPrecision(self,train = None,test = None,k = 8,nitem = 5):
        train = train or self.traindata
        test = test or self.testdata
        hit = 0
        recall = 0
        precision = 0
        users =[]
        for user in test.keys():
            tu = test.get(user,{})
            try:
                rank = self.recommend(user,train = train,k = k,nitem = nitem)
                for item,_ in rank.items():
                    if item in tu:
                        hit += 1
                recall += len(tu)
                precision += len(rank.keys())
            except KeyError:
                continue
        if recall ==0:
            recallrate=0
        else:
            recallrate= hit / (recall * 1.0) 
        if precision ==0:
            precisionrate=0
        else:
            precisionrate= hit / (precision * 1.0) 
        return (recallrate, precisionrate)
       
    def coverage(self,alls=None, train = None,test = None,k = 8,nitem = 5):
        train = train or self.traindata
        alls = alls or self.alldata
        recommend_items = set()
        all_items = set()
        for user in alls.keys():
            for item in alls[user].keys():
                all_items.add(item)
            rank = self.recommend(user, train, k = k, nitem = nitem)
            for item,_ in rank.items():
                recommend_items.add(item)
        
        return len(recommend_items) / (len(all_items) * 1.0)
    
    def popularity(self,alls=None,train = None,test = None,k = 8,nitem = 5):
        alls  = alls  or self.alldata
        train = train or self.traindata
        test  = test  or self.testdata
        item_popularity = dict()
        for user ,items in alls.items():
            for item in items.keys():
                item_popularity.setdefault(item,0)
                item_popularity[item] += 1
        ret = 0
        n = 0
        for user in test.keys():
            try:
                rank = self.recommend(user, train, k = k, nitem = nitem)
                for item ,_ in rank.items():
                    ret += math.log(1+item_popularity[item])
                    n += 1
            except KeyError:
                continue
        if n==0:
            popularity=0
        else:
            popularity=ret / (n * 1.0)
        return popularity
        
    def testRecommend(self,user):
        rank = self.recommend(user,k = 10,nitem = 5)
        for i,rvi in rank.items():
            items = self.traindata.get(user,{})
            record = items.get(i,0)
            print ("%5s: %.4f--%.4f" %(i,rvi,record))
                                
```
